```python
import pandas as pd

df = pd.read_csv('/content/gdrive/MyDrive/Dataset/EconomiesOfScale.csv')
```

```python
# Import library Pandas
import pandas as pd
```

```python
# Tampilkan 5 baris pertama dari DataFrame
print(df.head())
```

```python
# Tampilkan informasi dasar tentang DataFrame
print(df.info())
```

```python
# Isi kolom ini dengan Manufacturing Cost dibagi dengan Number of Units
df['Cost per Unit'] = df['Manufacturing Cost'] / df['Number of Units']

# Tampilkan DataFrame untuk memastikan kolom baru telah ditambahkan dengan benar
print(df.head())
```

```python
import matplotlib.pyplot as plt
```

```python
# Membuat boxplot
plt.figure(figsize=(12, 6))  # Atur ukuran plot

# Boxplot untuk Number of Units
plt.subplot(1, 2, 1)  # 1 baris, 2 kolom, subplot pertama
plt.boxplot(df['Number of Units'])
plt.title('Boxplot of Number of Units')
plt.xlabel('Number of Units')

# Boxplot untuk Manufacturing Cost
plt.subplot(1, 2, 2)  # 1 baris, 2 kolom, subplot kedua
plt.boxplot(df['Manufacturing Cost'])
plt.title('Boxplot of Manufacturing Cost')
plt.xlabel('Manufacturing Cost')

plt.tight_layout()  # Atur layout agar subplot tidak saling tumpang tindih
plt.show()  # Tampilkan plot
```

```python
# Buat plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Number of Units'], df['Manufacturing Cost'], c='blue', marker='o')
plt.title('Number of Units vs Manufacturing Cost')
plt.xlabel('Number of Units')
plt.ylabel('Manufacturing Cost')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
```

```python
# Buat plot
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Number of Units', y='Manufacturing Cost', data=df)
plt.title('Number of Units vs Manufacturing Cost')
plt.show()
```

```python
import numpy as np
```

```python
# Konversi kolom 'Cost per Unit' ke NumPy array
cost_per_unit = df['Cost per Unit'].to_numpy()

# Dapatkan indeks dari data yang telah diurutkan
sorted_indices = np.argsort(cost_per_unit)

# Urutkan seluruh DataFrame berdasarkan indeks yang telah diurutkan
df_sorted = df.iloc[sorted_indices]

# Tampilkan DataFrame yang telah diurutkan
print(df_sorted)
```
